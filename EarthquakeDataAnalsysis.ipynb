#Earthquake Data Analysis

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv('/content/Earthquake_Data.csv')
df.head()
```

```python
df.tail()
```

```python
df.describe()
```

#Data Visualization

```python
# Check if 'Country Name' and 'Number of Death' columns exist
if 'Country Name' in df.columns and 'Number of Death' in df.columns:
    # Create the chart
    plt.figure(figsize=(12, 6))  # Adjust figure size for better readability
    sns.countplot(x='Country Name', data=df, order=df['Country Name'].value_counts().index)
    plt.xticks(rotation=90)  # Rotate x-axis labels for better readability
    plt.xlabel("Country Name")
    plt.ylabel("Number of Deaths")
    plt.title("Number of Deaths by Country")
    plt.show()
else:
    print("Error: 'Country Name' or 'Number of Deaths' column not found in the DataFrame.")
```

```python
if 'Country Name' in df.columns and 'Earthquake Power' in df.columns:
    plt.figure(figsize=(12, 6))
    sns.boxplot(x='Country Name', y='Earthquake Power', data=df)
    plt.xticks(rotation=90)
    plt.xlabel("Country Name")
    plt.ylabel("Earthquake Magnitude")
    plt.title("Earthquake Magnitude by Country")
    plt.show()
else:
    print("Error: 'Country Name' or 'Magnitude' column not found in the DataFrame.")
```

```python
if 'Country Name' in df.columns and 'Earthquake Power' in df.columns and 'Number of Death' in df.columns:
    plt.figure(figsize=(12, 6))
    sns.scatterplot(x='Earthquake Power', y='Number of Death', hue='Country Name', data=df)
    plt.xlabel("Earthquake Power")
    plt.ylabel("Number of Deaths")
    plt.title("Earthquake Power vs. Number of Deaths by Country")
    plt.xticks(rotation=45)
    plt.legend(bbox_to_anchor=(1.05, 1), loc='upper left') #Move legend outside the plot
    plt.tight_layout() #Adjust layout to prevent labels from overlapping
    plt.show()
else:
    print("Error: 'Country Name', 'Earthquake Power', or 'Number of Deaths' column not found in the DataFrame.")
```
