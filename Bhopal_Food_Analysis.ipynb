#Bhopal Food Analysis

##Importing Libraries

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

##Importing dataset

```python
df = pd.read_csv('/content/Bhopal_Food_Price.csv')
```

```python
df.head()
```

```python
df.tail()
```

##Data Cleaning

```python
df.isnull().sum().sum()
```

```python
df.duplicated().sum()
```

##Data Visualization

```python
# Group data by year and calculate the mean price for rice
rice_prices_by_year = df.groupby('Year')['Rice'].mean()

# Create a bar graph
plt.figure(figsize=(10, 6))  # Adjust figure size as needed
plt.bar(rice_prices_by_year.index, rice_prices_by_year.values)
plt.xlabel('Year')
plt.ylabel('Average Rice Price')
plt.title('Average Rice Prices by Year')
plt.xticks(rotation=45)  # Rotate x-axis labels for better readability
plt.tight_layout()
plt.show()
```

```python
# Group data by year and calculate the mean price for bananas
wheat_prices_by_year = df.groupby('Year')['Wheat Flour (Atta)'].mean()

# Create a bar graph
plt.figure(figsize=(10, 6))  # Adjust figure size as needed
plt.bar(wheat_prices_by_year.index, wheat_prices_by_year.values)
plt.xlabel('Year')
plt.ylabel('Average Wheat floor Price')
plt.title('Average Wheat floor Prices by Year')
plt.xticks(rotation=45)  # Rotate x-axis labels for better readability
plt.tight_layout()
plt.show()
```

```python
# Group data by year and calculate the mean price for Lentils (Dal)
lentils_prices_by_year = df.groupby('Year')['Lentils (Dal)'].mean()

# Create a bar graph
plt.figure(figsize=(10, 6))  # Adjust figure size as needed
plt.bar(lentils_prices_by_year.index, lentils_prices_by_year.values)
plt.xlabel('Year')
plt.ylabel('Average Lentils (Dal) Price')
plt.title('Average Lentils (Dal) by Year')
plt.xticks(rotation=45)  # Rotate x-axis labels for better readability
plt.tight_layout()
plt.show()
```

```python
#Heat Map
correlation_matrix = df.corr()

plt.figure(figsize=(12, 8))
sns.heatmap(correlation_matrix, annot=True, cmap='coolwarm', fmt=".2f")
plt.title('Correlation Matrix of Food Prices')
plt.show()
```

##Inflation Rates

```python
# Calculate inflation rates
inflation_rates = df.pct_change() * 100

# Analyze inflation rates
print(inflation_rates.describe())
```

```python
# Calculate the inflation rate for rice prices
rice_inflation = df.groupby('Year')['Rice'].mean().pct_change() * 100

# Create a line chart
plt.figure(figsize=(10, 6))
plt.plot(rice_inflation.index, rice_inflation.values)
plt.xlabel('Year')
plt.ylabel('Rice Price Inflation (%)')
plt.title('Rice Price Inflation by Year')
plt.grid(True)
plt.show()
```

```python
# Calculate the inflation rate for wheat flour prices
wheat_inflation = df.groupby('Year')['Wheat Flour (Atta)'].mean().pct_change() * 100

# Create a line chart
plt.figure(figsize=(10, 6))
plt.plot(wheat_inflation.index, wheat_inflation.values)
plt.xlabel('Year')
plt.ylabel('Wheat Flour Price Inflation (%)')
plt.title('Wheat Flour Price Inflation by Year')
plt.grid(True)
plt.show()
```

```python
# Calculate the inflation rate for wheat flour prices
lentils_inflation = df.groupby('Year')['Lentils (Dal)'].mean().pct_change() * 100

# Create a line chart
plt.figure(figsize=(10, 6))
plt.plot(lentils_inflation.index, lentils_inflation.values)
plt.xlabel('Year')
plt.ylabel('Lentils (Dal) Price Inflation (%)')
plt.title('Lentils (Dal) Price Inflation by Year')
plt.grid(True)
plt.show()
```

##Recommended Minimum Amount of Money for food (2400 calories, Western food types)

```python
df1 = pd.read_csv('/content/Western_Food_Type.csv')
```

```python
df1.head()
```

```python
df1.tail()
```

```python
# Assuming 'Item' and 'Price' are columns in your df1 DataFrame
item_counts = df1.groupby('Item')['Price (₹)'].sum()

# Create the pie chart
plt.figure(figsize=(8, 8))
plt.pie(item_counts, labels=item_counts.index, autopct='%1.1f%%', startangle=90)
plt.title('Recommended Minimum Amount of Money for food (2400 calories, Western food types)')
plt.axis('equal')  # Equal aspect ratio ensures that pie is drawn as a circle.
plt.show()
```
