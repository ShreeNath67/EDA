#Super Market Sales Analysis

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv('/content/supermarket_sales.csv')
df.head()
```

```python
gender_counts = df['Gender'].value_counts()

plt.figure(figsize=(8, 6))
plt.pie(gender_counts, labels=gender_counts.index, autopct='%1.1f%%', startangle=90)
plt.title('Distribution of Gender')
plt.axis('equal')  # Equal aspect ratio ensures that pie is drawn as a circle.
plt.show()
```

```python
plt.figure(figsize=(8,6))
sns.countplot(x='Payment', data=df)
plt.title('Payment Method Counts')
plt.xlabel('Payment Method')
plt.ylabel('Count')
plt.show()
```

```python
product_gender = df.groupby(['Product line', 'Gender'])['Quantity'].sum().reset_index()

plt.figure(figsize=(12, 6))
sns.barplot(x='Product line', y='Quantity', hue='Gender', data=product_gender)
plt.title('Product line by Gender')
plt.xlabel('Product Line')
plt.ylabel('Total Quantity')
plt.xticks(rotation=45, ha='right')
plt.show()
```

```python
city_product = df.groupby(['City', 'Product line'])['Quantity'].sum().reset_index()

plt.figure(figsize=(12, 6))
sns.barplot(x='City', y='Quantity', hue='Product line', data=city_product)
plt.title('City by Product line')
plt.xlabel('City')
plt.ylabel('Total Quantity')
plt.xticks(rotation=45, ha='right')
plt.show()
```

```python
city_product = df.groupby(['City', 'Product line'])['Quantity'].sum().reset_index()

plt.figure(figsize=(12, 6))
sns.barplot(x='City', y='Quantity', hue='Product line', data=city_product)
plt.title('City by Product line')
plt.xlabel('City')
plt.ylabel('Total Quantity')
plt.xticks(rotation=45, ha='right')
plt.show()
```
