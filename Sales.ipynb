#Sales data


```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv('/content/1000 Sales Records.csv')
df.head()
```

```python
df.tail()
```

```python
df.isnull().sum().sum()
```

```python
df.duplicated().sum().sum()
```

```python
plt.figure(figsize=(16, 6))
sns.countplot(x='Country', data=df)
plt.xticks(rotation=90)
plt.show()
```

```python
sales_by_channel = df['Sales Channel'].value_counts()

plt.figure(figsize=(8, 8))
plt.pie(sales_by_channel, labels=sales_by_channel.index, autopct='%1.1f%%', startangle=90)
plt.title('Sales by Channel')
plt.axis('equal')  # Equal aspect ratio ensures that pie is drawn as a circle.
plt.show()
```

```python
top_10_countries = df['Country'].value_counts().nlargest(10).index

plt.figure(figsize=(12, 6))
sns.countplot(x='Country', data=df[df['Country'].isin(top_10_countries)], hue='Item Type')
plt.xticks(rotation=45)
plt.title('Top 10 Countries by Item Type')
plt.show()
```

```python
top_10_countries = df['Country'].value_counts().nlargest(10).index

plt.figure(figsize=(12, 6))
sns.barplot(x='Country', y='Units Sold', hue='Item Type', data=df[df['Country'].isin(top_10_countries)], estimator=sum)
plt.xticks(rotation=45)
plt.title('Top 10 Countries by Units Sold and Item Type')
plt.xlabel('Country')
plt.ylabel('Total Units Sold')
plt.tight_layout() # Adjust layout to prevent labels from overlapping
plt.show()
```
